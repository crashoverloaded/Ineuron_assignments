```python
# Standard Procedure for Logging and Debugging of Code
# DEBUG -> Detail information about the process
# INFO -> Information about software
# WARNING -> Warnings about installations
# ERROR
# CRITICAL
```

```python
import logging as lg
```

```python
# Storing in test.log file and all the INFO will be saved which is specified using level parameter
# using level parameter we can specify the type of log you want to save , such as INFO , DEBUG , etc....
# After thhis we can try to change format of the log , in which we can show timestamp and message using format parameter
lg.basicConfig(filename = 'test.log', level = lg.INFO, format ="%(asctime)s %(message)s")
```

```python
# Writing to the INFO file which we have created 
lg.info("I am going to start my program")
```

```python
# Writing warning to the file
lg.warning("This is first warning of my program")
```

```python
# Writing error in the file
lg.error("This is error message")
```

```python
# Shutting down logging file
lg.shutdown()
```

```python
# New logging example
# basically in this example we are trying to access logs from different handlers or users , hence in the output 
# Given below we can see that logger1.area has different logs and logger2.area has different logs
import logging as lg
import sys
# Defining Level as Debug
lg.basicConfig(filename="app.log",level = lg.DEBUG,format = '%(name)s - %(asctime)s - %(levelname)s - %(message)s')

# we can implement and set a level using streamhandle
# Printing  all the log into console can be done by below code
console_log = lg.StreamHandler()

# This statement allows to show only DEBUG log to the console
console_log.setLevel(lg.DEBUG)

# Changing format of log for console_log variable 
format = '%(name)s - %(asctime)s - %(levelname)s - %(message)s'
console_log.setFormatter(format)

# Adding different kinds of handlers
# a handler can be different processes creating different logs , Hence to seperate loggers like logger
# we can add handlers to the console_log
lg.getLogger('').addHandler(console_log)
lg.info('main log')
# Adding different loggers
# we can write anything like usernames and process , etc..
logger1 = lg.getLogger('logger1.area1')
logger2 = lg.getLogger('logger2.area2')
user1 = lg.getLogger('user1')
# Now addings logs to different loggers
logger1.info("INFO log of the first logger")
logger2.debug("Debug log for the second logger")
logger2.info("INFO log for second logger")
user1.info("INFO of user1")
user1.debug("DEBUG of user1")
```

# DEBUGGING EXAMPLE


```python
import ipdb
```

```python
# set_trace can be used to check the flow of the code
def testdebug():
    ipdb.set_trace()
    l = []
    for i in range(10):
        l.append(i)
        print("Appending is done")
    return l
```

```python
testdebug()
```
